```python
import re

def clean_text(text):
    if not isinstance(text, str):
        return ""
        # 1
    text = re.sub(r'<[^>]+>','', text)
        # 2
    text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)
    text = re.sub(r'\S*@\S*\s?', '', text)
        # 3
    vietnamese_chars = "a-zA-ZàáảãạăằắẳẵặâầấẩẫậèéẻẽẹêềếểễệìíỉĩịòóỏõọôồốổỗộơờớởỡợùúủũụưừứửữựỳýỷỹỵđÀÁẢÃẠĂẰẮẲẴẶÂẦẤẨẪẬÈÉẺẼẸÊỀẾỂỄỆÌÍỈĨỊÒÓỎÕỌÔỒỐỔỖỘƠỜỚỞỠỢÙÚỦŨỤƯỪỨỬỮỰỲÝỶỸỴĐ"
    pattern = f"[^{vietnamese_chars}\d\s.,?!:;]"
    text = re.sub(pattern, '', text)
        # 4
    text = re.sub(r'\s([.,?!:;])', r'\1', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text
```

```python
# use secrete key
from google.colab import userdata
from firecrawl import Firecrawl
FIRECRAWL_API = userdata.get('FIRECRAWL')
app = Firecrawl(api_key=FIRECRAWL_API)
urls = ["https://vnexpress.net/gia-vang-moi-nhat-hom-nay-ngay-23-12-4997154.html","https://vnexpress.net/han-quoc-nhat-ban-tranh-ve-chung-ket-u23-chau-a-5007701-tong-thuat.html" ]
# content_firecrawl = app.scrape(url=url, formats=["markdown"])
```

```python
ROOT_FOLDER ="/content/drive/MyDrive/Project_Datacom"
SAVE_FILE = "article.txt"
```

```python
for url in urls:
    content_firecrawl = app.scrape(url=url, formats=["markdown"]).markdown
    with open(f"{ROOT_FOLDER}/{SAVE_FILE}", "a", encoding="utf-8") as f:
        f.write(f"{content_firecrawl}\n")
```

```python
with open(f"{ROOT_FOLDER}/{SAVE_FILE}", "r", encoding="utf-8") as f:
    content = f.read()
```

```python
content
```

```python
cleaned_article = clean_text(content)
cleaned_article
```
